```python
import pandas as pd
frame = pd.read_excel('學生第一學期分數.xlsx')
#frame1 = frame.rename({'Unnamed: 0':'學號'})
frame
```

```python
frame2 = frame[['學號','國文','英文','數學','地理','歷史']]
frame3 = frame2.set_index('學號')
frame3
```

```python
def description(s:pd.Series) -> pd.Series:
    max = s.max()
    max_index = s[s==max].index[0]
    min = s.min()
    min_index = s[s==min].index[0]
    std = s.std()
    return pd.Series([max_index,max, min_index,min, std],index=['最高分科目','最高分','最低分科目','最低分','標準差'])

frame3.apply(description,axis=1)
```
